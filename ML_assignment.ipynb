5. Dropped the duplicate rows to avoid redundant data which makes the task easier

```python
unique_items_df = raw_skill_df.drop_duplicates(keep='last',ignore_index=True)
```

6. Transfered the raw data column values into a list

```python
skill_list= unique_items_df['RAW DATA'].to_list()
```

7. created a regular expression that searches for text data in the following format:

a. Start with one or more capital letters followed by word characters<br>
b. Ignoring a|an|0-9|:|,|.| |\n|\t5 from the text data<br>
c. Followed by blank and letters ranging from A-Z,a-z with one or more occurrences

```python
regex = r"^[A-Z]*\w[^an0-9:,. \n.\t5*]\s[A-Za-z]*"
```

8. Wrote a code to append the text data that follows the above regular expression

```python
final_list = []
for item in skill_list:
    pattern = re.search(regex,item)
    if pattern:
        final_list.append(item)
        
```

9. With the final list created a new skill dataframe

```python
new_skill_df = pd.DataFrame(data=final_list,columns=['Skills'])
```
